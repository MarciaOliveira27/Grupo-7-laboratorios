```python
#Pesquisar artigos sobre HNF4A
handle = Entrez.esearch(db="pubmed", sort='relevance', term='HNF4A')
record = Entrez.read(handle)
lista = record["IdList"]
```

```python
len(lista)

print(lista)
```

```python
#corresponder os records e extrair informação
from Bio import Medline
handle = Entrez.efetch(db='pubmed', id=lista, rettype='medline')
```

```python
records = Medline.parse(handle)
```

```python
records = list(records) #para guardar os records 
```

```python
for record in records:
    print("title:", record.get("TI","?"))
    print("authors:", record.get("AU","?"))
    print("source:", record.get("SO","?"))
    print('')
```

```python
print(records[0].keys())
```

```python
handle = Entrez.egquery(term='HNF4A')
record = Entrez.read(handle)
for row in record['eGQueryResult']:
    if row['DbName']=='nuccore':
        print(row['Count'])
```

```python
#GENBANK identifiers
handle = Entrez.esearch(db='nuccore', term="HNF4A")
record = Entrez.read(handle)
gi_list = record["IdList"]
gi_list
```

```python
gi_str = ','.join(gi_list)
handle = Entrez.efetch(db='nuccore', id=gi_str, rettype = 'gb', retmode='text', retmax=[10])
#para fazer o download dos records GenBank (passar para uma lista que depois é convertida)
```
