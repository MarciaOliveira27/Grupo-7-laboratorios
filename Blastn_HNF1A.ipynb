```python
from Bio.Blast import NCBIXML 
from Bio.Blast import NCBIWWW 
from Bio import SeqIO
```

```python
record = SeqIO.read("GenBank_HNF1A.gb", "genbank")
```

```python
features = ((record.features),"features") #tipo e localização de features
```

```python
for feat in record.features:
    print("-->", feat)
print(f"Número de features: {features}")
for feat in record.features:
    print("Type:", feat.type)
    print("Location:", feat.location)
```

```python
featcds = [ ]
for i in range(len(record.features)):
    if record.features[i].type == "CDS":
        featcds.append(i)
for k in featcds:
    print (record.features[k].location)
for k in featcds:
    print (record.features[k].extract(record.seq))
print(featcds)

for feat in record.features:
    if feat.type == 'CDS':
        print("Proteína codificada: ", feat.qualifiers['product'])

for feat in record.features:
    if feat.type == 'gene':
        print("Significado biológico: ", feat.qualifiers["note"])
```

```python
records = SeqIO.parse("GenBank_HNF1A.gb","genbank")
count = SeqIO.write(records, "HNF1A_fasta.fa", "fasta")
print(f'Foi convertido {count} registo.')
```

```python
record = SeqIO.read("HNF1A_fasta.fa", format="fasta")
result_handle = NCBIWWW.qblast("blastn", "nt", record.seq)
```

```python
save_file = open("HNF1A_blastn.xml", "w")
save_file.write(result_handle.read())
save_file.close()
result_handle.close()
```

```python
result_handle = open("HNF1A_blastn.xml")
```

```python
blast_record = NCBIXML.parse(result_handle)
for b in blast_record:
    print(f"Database: {b.database}")
    print(f"Matriz de Substituição: {b.matrix}")
    print(f"Gap penalty: {b.gap_penalties}") 
```
