```python
#Identificação de features do tipo CDS e a sua localização:
featcds = [] 
for i in range(len(seq_record.features)):
    if seq_record.features[i].type=='CDS':
        featcds.append(i)
for k in featcds:
    print("Features do tipo CDS e a sua localização:\n",seq_record.features[k].location)
```

```python
#Sub-sequência de DNA afetada pela feature:
for k in featcds:
    print("sub-sequência de DNA afetada pela feature:\n",seq_record.features[k].extract(seq_record.seq))
```
