__Blastp P20823 (HNF1A)__

```python
from Bio.Blast import NCBIWWW
from Bio.Blast import NCBIXML
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
import re
```

```python
P20823_record = SeqIO.read(open('P20823.fasta'),format='fasta') #obter o handle com os resultados usando o blastp
P20823_result = NCBIWWW.qblast('blastp','nr',P20823_record.format('fasta'), hitlist_size=20)
```

```python
save_file = open("blast_P20823.xml", "w")
save_file.write(P20823_result.read())
save_file.close()
P20823_result.close()
```

```python
P20823_result = open("blast_P20823.xml") #ler resultados do blast
```

```python
org = []
HSP = []
ID = []
evalue_tresh = 1e-20
blast_record = NCBIXML.read(P20823_result)
for alignment in blast_record.alignments:
    for hsp in alignment.hsps:
        if hsp.expect < evalue_tresh:
            organismo =  re.findall(r'([\w\s]+)]', alignment.title)
            org.append(organismo[0])
            HSP.append(hsp.sbjct)
            ID.append(alignment.hit_id)
            print("      ***ALINHAMENTO***")
            print("Sequência: ", alignment.title)
            print("Tamanho da sequência: ", alignment.length)
            print("E-value:", hsp.expect)
            print("Score: ", hsp.score)
            print(f"Caracteres iguais: {len(hsp.match)}")
            print(hsp.query[0:100] + "...")
            print(hsp.match[0:100] + "...")
            print(hsp.sbjct[0:100] + "...")
```

```python
ID_dif = []
HSP_dif = []
org_dif = []
for x in range(len(org)):
    if org[x] not in org_dif:
        ID_dif.append(ID[x])
        org_dif.append(org[x])
        HSP_dif.append(HSP[x])
```

```python
org_diff = []
for x in range(len(org_dif)):
    org_diff.append(org_dif[x].replace(" ", "_"))
```

```python
blast_P20823 = open("seqs_P20823.fasta", "w") ##converter o ficheiro xml do blastp para fasta
for x in range(len(org_diff)):
        blast_P20823.write(">" + org_diff[x] + "\n"+ HSP_dif[x] + "\n")
```
